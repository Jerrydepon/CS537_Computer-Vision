```python
import cv2
import numpy as np
import os
import torch
import random
import matplotlib.pyplot as plt


# def writeTxtFile(fileName, keypoints):
#     f = open("{0}.txt".format(fileName), "a")
#     for (x, y) in keypoints:
#         f.write("{0}, {1}\n".format(x, y))
#     f.close()


def getPatches(kps, img, size=32, num=500):
    res = torch.zeros(num, 1, size, size)
    if type(img) is np.ndarray:
        img = torch.from_numpy(img)
    h, w = img.shape  # note: for image, the x direction is the verticle, y-direction is the horizontal...
    for i in range(num):
        cx, cy = kps[i]
        cx, cy = int(cx), int(cy)
        dd = int(size / 2)

        xmin, xmax = max(0, cx - dd), min(w, cx + dd) - 1
        ymin, ymax = max(0, cy - dd), min(h, cy + dd) - 1
        xmin_res, xmax_res = dd - min(dd, cx), dd + min(dd, w - cx) - 1
        ymin_res, ymax_res = dd - min(dd, cy), dd + min(dd, h - cy) - 1

        res[i, 0, ymin_res: ymax_res, xmin_res: xmax_res] = img[ymin: ymax, xmin: xmax]
#         print(ymin, ymax, xmin, xmax)
    return res


def getInterestPoints(image_dir):
    image = cv2.imread(image_dir)
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    sift = cv2.xfeatures2d.SIFT_create(200)
    (kps, descs) = sift.detectAndCompute(gray, None)
    return kps, gray

# keypoints = Array
def getKeypointsTorch(keypoints):
    rt = torch.zeros(200, 2)
    for i in range(0,200):
        (x,y)= keypoints[i]
        rt[i][0] = x
        rt[i][1] = y
    return rt
    
#     img = cv2.imread(image_dir)
#     gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
#     gray = np.float32(gray)
#     dst = cv2.cornerHarris(gray, 2, 21, 0.15)
#     return dst > 0.05 * dst.max(), gray


all_patches = torch.zeros(5, 200, 1, 32, 32)
all_keypoints = torch.zeros(5, 200, 2)
files, num = [], 0
img_dirs_Arr = ['images/NotreDame1.jpg', 
       'images/NotreDame2.jpg', 
       'images/NotreDame3.jpg',
       'images/NotreDame4.jpg', 
       'images/NotreDame5.jpg']
for img_dir in img_dirs_Arr:
    (kps, gray), tmp = getInterestPoints(img_dir), []
    for x in kps:
        tmp.append((x.pt[0], x.pt[1]))

    key_points = random.sample(tmp, 200)
    patches = getPatches(key_points, gray, size=32, num=len(key_points))
    all_patches[num] = patches
    all_keypoints[num] = getKeypointsTorch(keypoints=key_points)
    num += 1
    
#     if num < 2:
#         for patch in patches:
#             im = patch[0].numpy()
#             plt.imshow(im)
#             plt.show()
    
    
#     writeTxtFile(fileName=imgFileName.replace(".jpg", ""), keypoints=key_points)

output_patch_dir = "{0}.pt".format("wangchih_patchs")
output_keypoints_dir = "{0}.pt".format("wangchih_keypoints")

torch.save(all_keypoints, output_keypoints_dir)
test_keys = torch.load(output_keypoints_dir)
print(type(test_keys))
print("all keypoints: ", test_keys.shape)


torch.save(all_patches, output_patch_dir)
test_patches = torch.load(output_patch_dir)
print(type(test_patches))
print("all patches: ", test_patches.shape)
```
